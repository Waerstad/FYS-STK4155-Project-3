# Heat Equation Solver Test

```python
from HeatEqSolver import HeatEqSolver
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 1, 100)
initial_state = np.sin(np.pi*xs)
space_step = 0.01
time_step = 1/2*space_step**2
duration = 0.2
solver = HeatEqSolver(initial_state, space_step, time_step, duration)

print(solver._num_time_steps)
print(solver._num_space_points)
print(solver._init_solver_matrix().shape)

solution = solver.solve()
time_points = solution.shape[0]
color_init = np.array([1.0, 0.0, 0.0])
color_end = np.array([0.0, 0.0, 1.0])
for i, time_state in enumerate(solution):
    if i % 1000 == 0:
        alpha = 0.5 + 0.5*(i/time_points)
        color = color_init*(1-i/time_points) + color_end*(i/time_points)
        plt.plot(xs, time_state, color=color, alpha=1, label=r"$t=$"+f"{i*time_step:.2f}")
plt.legend()
plt.show()
```
