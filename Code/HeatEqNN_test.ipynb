# Neural Network Heat Equation Solver Test

```python
from HeatEqNN import HeatEqNN
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
nn = HeatEqNN(2, 20, "tanh", learning_rate=0.001)
```

```python
rng = np.random.default_rng()
xs = rng.random((100))*1
ts = rng.random((100))*0.3
#xs = np.linspace(0, 1, 100)
#ts = np.linspace(0, 0.2, 100)
#xs = np.geomspace(1, 1/100, 100)
#ts = np.geomspace(0.2,1/100,100)
xs_grid,ts_grid = np.meshgrid(xs,ts)
points = np.stack([xs_grid.ravel(), ts_grid.ravel()], axis=1)
num_points = np.shape(points)[0]
print(xs[:5])
print(ts[:5])
print(xs[-5:])
print(ts[-5:])
print("Points:", num_points)
```

```python
nn.train(points, epochs = 1000)
```

```python
space_n = 100
time_n = 5
xs_plot = np.linspace(0,1, space_n)
ts_plot = np.linspace(0,0.2,time_n)
```

```python
print(ts_plot)
for i, t in enumerate(ts_plot):
    pred_points = np.array([xs_plot, t*np.ones(space_n)]).T
    u_pred = nn.predict(pred_points)
    plt.plot(xs_plot, u_pred, label = r"$t = $" + f"{t:.2f}")
plt.legend()
plt.show()
```

```python
a = nn.test(pred_points)
print(a.numpy())
```

```python
## For comparison, define the analytical solution
def g_analytic(point):
    x,t = point
    return np.exp(-np.pi**2*t)*np.sin(np.pi*x)
```

```python
colors = mpl.color_sequences['tab20']
print(ts_plot)
for i, t in enumerate(ts_plot):
    pred_points = np.array([xs_plot, t*np.ones(space_n)]).T
    u_pred = nn.predict(pred_points)
    plt.plot(xs_plot, u_pred, label = r"NN $t = $" + f"{t:.2f}", color = colors[2*i])
    plt.plot(xs_plot, g_analytic(pred_points.T), linestyle="--", label = r"AN $t = $" + f"{t:.2f}", color = colors[2*i+1])
plt.legend()
plt.show()
```
